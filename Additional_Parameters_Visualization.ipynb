```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
main_df = pd.read_csv('parameters_graph_visualisation.csv')
main_df['name'] = main_df['Projects']
main_df = main_df[['name', 'Domain', 'Type', 'Organisation']]
main_df.head()
```

```python
clustering = pd.read_csv('clustering_result_30_ver.csv')
clustering.head()
```

```python
clustering = pd.merge(clustering, main_df, on='name')
clustering.head()
```

```python
clustering_domain_mean = clustering.groupby('Domain')['mojoFM'].mean()
clustering_domain_mean.plot.bar()
```

```python
clustering_domain_max = clustering.groupby('Domain')['mojoFM'].max()
clustering_domain_max.plot.bar()
```

```python
clustering_domain_min = clustering.groupby('Domain')['mojoFM'].min()
clustering_domain_min.plot.bar()
```

```python
clustering_domain_min = clustering.groupby(['Domain', 'affinity'])['mojoFM'].max()
fig = clustering_domain_min.plot.bar(figsize=(12,12))
```

```python
clustering_domain_min = clustering.groupby(['Type', 'affinity'])['mojoFM'].min()
fig = clustering_domain_min.plot.bar(figsize=(20,20))
```

```python
clustering_domain_min = clustering.groupby(['Organisation', 'affinity'])['mojoFM'].min()
fig = clustering_domain_min.plot.bar(figsize=(12,12))
```

```python
clustering_domain_min = clustering.groupby(['Organisation', 'n_clusters'])['mojoFM'].min()
fig = clustering_domain_min.plot.bar(figsize=(12,12))
```

```python
clustering_domain_min = clustering.groupby(['Organisation', 'affinity'])['mojoFM'].min()
fig = clustering_domain_min.plot.bar(figsize=(12,12))
```
