```python
full_dir_arr = []
for root, dirs, files in os.walk(rootdir):
    #print(root)
    #print(dirs)
    for element in files:
        if '.java' in element:
            dir_string = root + '\\' + element
            full_dir_arr.append(dir_string)
```

```python
cluster_dict = {}
cluster_tree = {}

    
for element in full_dir_arr:

    element = element.split('\\')
    i = len(element)
    while i > 1:
        #print(element[i-1])
        parent = element[i-2]
        child = element[i-1]
        #print(parent)
        #print('#')
        #print(child)
        i -= 1
        if parent in cluster_dict:
            cluster_dict[parent] += 1
        else:
            cluster_dict[parent] = 1

        cluster_tree[child] = parent

        
        
```

```python
total_clusters = 0
extreme_clusters = 0
for key in cluster_dict:
    total_clusters += 1
    if cluster_dict[key] < 5 or cluster_dict[key] > 100:
        extreme_clusters += 1

final_extremity_score = 1 - (extreme_clusters/total_clusters)
print(final_extremity_score)
```
